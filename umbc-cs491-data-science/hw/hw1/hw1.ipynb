### State

#### Description

This is the origin State for the license plate of the violation. There are no missing datapoints.

#### Outliers

Since this is a categorical value, I don't think there will be any outliers persay. I did want to get the number of unique values in the dataframe however. You can see that bit below.

```python
unique_states = df.groupby('State')['State'].count()
display(unique_states.sort_values(ascending=False))
```

#### Observations

Unsurprisingly, the most common state of vehicle to get tickets in Baltimore is Maryland. Next most common are Virginia and Pennsylvania, two states that border Maryland. I was surprised how many 1-ticket states there were. I was assuming there would be around 50 but there's actually 125! I assume a lot of those are Reservations which issue their own license plates.
