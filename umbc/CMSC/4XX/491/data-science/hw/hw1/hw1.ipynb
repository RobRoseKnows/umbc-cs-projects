However, its distribution will be quite similar to `ViolFine` because the remaining tickets will be paid off.

#### Outliers

While we found a few extreme values in `ViolFine`, we chose not to do anything with them because there were enough to be meaningful and removing them would remove a large amount of ticket classes.

#### Observations

As predicted, there's a lot more `OpenFine`'s equal to zero because of people paying off the fines. Unfortunately it looks like many people don't pay off high value tickets. I wonder if that means they're fighting them in court or something else.

### OpenPenalty

The `OpenPenalty` is the amount owed on a ticket that's overdue and has been asigned a penalty. I expect for a lot of tickets this will be zero.

Wow it looks like some people owe more than 1000 to the city. I checked to see what tickets people owe this much on and the results were pretty spread out. It might be worthwhile dropping outliers.

```python
print(df['Description'].loc[df['OpenPenalty'] > 1000].unique())
print(df['Description'].loc[df['OpenPenalty'] > 1000].count())
print(df['Description'].loc[df['OpenPenalty'] > 2000].unique())
print(df['Description'].loc[df['OpenPenalty'] > 2000].count())
```

Since there was only one `OpenPenalty` greater than 2000 with a common violation type, I decided to drop it to make the data cleaner.

#### Observations

Overall the `OpenPenalty` column followed my prediction of there being lots of 0's and a few very high penalties.

## Scatter Plots

In this section I do some scatter plotting to find correlation. The first thing I did however was use Pandas to create a correlation matrix to give me an idea of where I should look for correlation.

```python
corr = df.corr()
corr.style.background_gradient()
```

It looks like `OpenFine`, `OpenPenalty` and `Balance` are all highly correlated with one another. This is fairly expected as `Balance` is essentially a sum of `OpenFine` and `OpenPenalty`. Lets plot them. In order to make plotting the data a realistic goal, I limited it to the first 50,000 items.

### OpenPenalty -> Balance

I think these two variables are related because the data labels suggests they are and the correlation grid shows an extremely high correlation betwen the two. The plot shows an extremely strong correlation between OpenPenalty and Balance because it looks like most people's Balances are taken up by penalties, not the fines themselves.

```python
sns.jointplot("OpenPenalty", "Balance", data=df[:50000], kind="reg", height=7)
```

### OpenFine -> OpenPenalty

I thought these two might be related earlier and the correlation grid confirmed my suspicions. The plot shows that there is a very small positive correlation between `OpenFine` and `OpenPenalty`. I assume this is because if someone gets a more expensive ticket, they are less likely to pay it immediately and therefore may get a penalty.

```python
sns.jointplot("OpenFine", "OpenPenalty", data=df[:50000], kind="reg", height=7)
```

## Predictor Plot

```python
sns.scatterplot(x="OpenFine", y="OpenPenalty", size="Balance", sizes=(10,200), data=df[:50000])
```

I'm fairly confident `OpenFine` and `OpenPenalty` are predictor variables for `Balance` because `Balance` is basically just the sum of `OpenFine` and `OpenPenalty`. You can also see this on the scatter plot, the `Balance`'s are very small in the bottom left when `OpenFine` and `OpenPenalty` are close to zero and very large as `OpenFine` and `OpenPenalty` increase.
